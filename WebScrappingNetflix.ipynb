```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.netflix.com/br/browse/genre/34399'
```

```python
requisicao = requests.get(url)
```

```python
requisicao.status_code
```

```python
pagina = requisicao.content
```

```python
type(pagina)
```

```python
pagina = BeautifulSoup(pagina , 'html.parser')
```

```python
pagina.find('ul', attrs = {'class': 'nm-content-horizontal-row-item-container'})
```

```python
ul = pagina.find('ul', attrs = {'class': 'nm-content-horizontal-row-item-container'})
```

```python
print(ul.prettify())
```

```python
filmes =[]
hrefs = []
for cada_li in ul.findAll('li'):
  tag_a = cada_li.find('a')
  hrefs.append(tag_a.get('href'))
  span = tag_a.find('span' , attrs = {'class':'nm-collections-title-name'})
  if span:
    nome_filme = span.text
    filmes.append(nome_filme)


print(filmes)
print(hrefs)
```

```python
filmes_links = dict(zip(filmes , hrefs))
filmes_links
```

```python
df = pd.DataFrame({'Filmes': filmes , 'Link': hrefs})
```
